```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
dataset = keras.preprocessing.image_dataset_from_directory(
  '/content/drive/MyDrive/train (1)/', batch_size=72,image_size=(256, 256), shuffle=True, seed = 210,
)
```

```python
import numpy as np
import pandas as pd
import os
from keras import applications
from tensorflow.keras.applications import ResNet50
from keras import callbacks
import tensorflow as tf
from keras.models import Sequential
```

```python
mod = tf.keras.applications.ResNet50(
    include_top=False,
    weights="imagenet",
    input_shape=(256,256,3),
    pooling=None,
    classes=1000,
    classifier_activation="softmax",
)
for layer in mod.layers:
    layer.trainable = False
mod.layers[-2].trainable = True

```

```python
from tensorflow.keras.models import *
from tensorflow.keras.layers import *
model = Sequential()
model.add(Input(shape=(256,256,3)))
model.add(Rescaling(1./255))
model.add(mod)
model.add(Conv2D(32, kernel_size=(3,3),padding="same",input_shape=(256,265,3),activation="relu"))
model.add(MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))
model.add(Conv2D(10, kernel_size=(2,2), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(11, activation='softmax'))
```

```python
callback = tf.keras.callbacks.EarlyStopping( patience=10,
                                          min_delta=0.001,
                                         )
```

```python
model.compile(
  optimizer="adam", metrics=['acc'],
  loss='sparse_categorical_crossentropy',
 )
```

```python
history = model.fit(dataset, epochs=6,callbacks=[callback])
```

```python

test_path = "/content/drive/MyDrive/public_test"
```

```python
predictions =[]
for i in range(1436):
    img = tf.keras.preprocessing.image.load_img(test_path +'/'+ str(i)+'.jpg')
    img = tf.keras.preprocessing.image.img_to_array(img)
    img = tf.keras.preprocessing.image.smart_resize(img, (256, 256))
    img = tf.reshape(img, (-1, 256, 256, 3))
    prediction = model.predict(img)
    predictions.append(np.argmax(prediction))
```

```python
colors = ['Black',
 'Blue',
 'Brown',
 'Cyan',
 'Green',
 'Grey',
 'Orange',
 'Red',
 'Violent',
 'White',
 'Wellow']
```

```python
sub = []
for i in range(len(predictions)):
  sub.append(colors[predictions[i]])
```

```python
my_submission = pd.DataFrame([*zip(sub)])
my_submission.to_csv('submission3.csv', index=False)
```
